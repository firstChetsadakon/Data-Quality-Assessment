Vailidity Check
* ทุก Column ถ้าข้อมูลเป็น Missing Value : Validity จะต้องเป็น Nan
* Transaction ID เช็คว่าขึ้นต้นด้วย TXN_{เลข 7 ตัวหรือไม่}
* Customer ID เช็คว่า CUST_{เลขกี่ตัวก็ได้}
* Category เช็คว่าอยู่ใน list นี้หรือไม่['Patisserie','Milk Products','Butchers','Beverages','Food','Furniture','Electric household essentials','Computers and electric accessories']
* Item เช็คว่าเขียนอยู่ใน pattern Item_{เลขกี่ตัวก็ได้}_{ตัวหนังสือกี่ตัวก็ได้} หรือ ไม่ ถ้าไม่จะเป็น False 
    * ถ้าใช่ดูว่า Column Category มีข้อมูลหรือไม่ถ้าไม่ให้นับว่า CANT CHECK ถ้า มี Category ต้องดูว่า เขียนถูก pattern Item_{เลขกี่ตัวก็ได้}_{ตัวย่อของ CAt} ไหม ถ้าถูก จะเป็น True ผิดจะ False
    * ถ้าเขากรอก Category มาผิด (ไม่อยู่ใน List ด้านบน) จะขึ้นว่า CANT CHECK
* Price Per Unit เช็คว่า Price เป็นเลขหรือไม่ ถ้าไม่ใช่นับว่าเป็น False ถ้าใช่ให้ดูว่า Price ใน Transaction Date ตรงกันกับ ราคาของ Item หรือเปล่าจาก Price Master ถ้าตรงเป็น True ไม่ตรง False
    * ถ้า Item หรือ Transaction Date ผิด ให้นับว่าเป็น CANT CHECk
* Quantity ต้องเป็นเลข จำนวนเต็ม ไม่ 0 ไม่ติดลบ 
* Total Spent = QTY * Price ถ้า price หรือ QTY หายหรือผิดให้เป็น CANT CHECK 
* Payment Method ต้องอยู่ใน list ['Digital Wallet', 'Credit Card', 'Cash']
* Location ต้องอยู่ใน list ['Online', 'In-store']
* Transaction Date ต้องอยู่ใน format '2024-01-31' (ปี-เดือน-วัน) และต้องเป็น date ที่ถูกต้อง (ไม่ใช่ 32 มกรา ไม่ใช่ 30 กุมภา)
* Discount Applied ต้องเป็น True หรือ False
