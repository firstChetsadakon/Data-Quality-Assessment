```python
import re

def validate_column_format(df, column_name, regex_pattern):
    """
    ฟังก์ชันตรวจสอบว่า:
    1. ค่าภายในคอลัมน์มี Missing Values กี่เปอร์เซ็นต์
    2. ค่าที่ไม่เป็น Missing Values ตรงกับ Format ที่ระบุหรือไม่ และคำนวณเปอร์เซ็นต์ความถูกต้อง

    Parameters:
        df (pd.DataFrame): DataFrame ที่ต้องการตรวจสอบ
        column_name (str): ชื่อคอลัมน์ที่ต้องการตรวจสอบ
        regex_pattern (str): Regular Expression (Regex) ที่ใช้ตรวจสอบ Format

    Returns:
        missing_percentage (float): เปอร์เซ็นต์ข้อมูลที่หายไป (NaN หรือค่าว่าง)
        valid_percentage (float): เปอร์เซ็นต์ข้อมูลที่ไม่หายและตรงกับ Format
        df (pd.DataFrame): DataFrame พร้อมคอลัมน์ Valid ที่บอกว่าข้อมูลผ่านการตรวจสอบหรือไม่
    """

    # ตรวจสอบค่าที่เป็น Missing (NaN หรือค่าว่าง)
    missing_values = df[column_name].isna().sum() + df[column_name].astype(str).apply(lambda x: x.strip() == "").sum()
    missing_percentage = (missing_values / len(df)) * 100

    # ตรวจสอบเฉพาะค่าที่ไม่หายไป (Non-Missing)
    df_valid = df[df[column_name].notna() & (df[column_name].astype(str).str.strip() != "")]
    df_valid[f"Valid_{column_name}"] = df_valid[column_name].astype(str).apply(lambda x: bool(re.fullmatch(regex_pattern, x)))

    # คำนวณเปอร์เซ็นต์ค่าที่ไม่หายและตรง Format
    total_non_missing = len(df_valid)
    valid_count = df_valid[f"Valid_{column_name}"].sum()
    valid_percentage = (valid_count / total_non_missing) * 100 if total_non_missing > 0 else 0

    # แสดงผลลัพธ์
    print(f"✅ คอลัมน์ '{column_name}' สถานะ:")
    print(f"   - ข้อมูลที่หายไป: {missing_values} จากทั้งหมด {len(df)} ({missing_percentage:.2f}%)")
    print(f"   - ข้อมูลที่ไม่หายและตรง Format: {valid_count} จาก {total_non_missing} ({valid_percentage:.2f}%)")

    return missing_percentage, valid_percentage, df
```

```python
import re
from datetime import datetime

def validate_transaction_date(df, column_name):
    """
    ฟังก์ชันตรวจสอบว่า:
    1. ค่าภายในคอลัมน์มี Missing Values กี่เปอร์เซ็นต์
    2. ค่าที่ไม่เป็น Missing Values ต้องอยู่ใน Format YYYY-MM-DD หรือไม่
    3. ตรวจสอบว่า:
       - ปีต้องมากกว่า 1800
       - เดือนต้องไม่เกิน 12
       - วันต้องถูกต้องตามเดือน (รองรับ Leap Year)

    Parameters:
        df (pd.DataFrame): DataFrame ที่ต้องการตรวจสอบ
        column_name (str): ชื่อคอลัมน์ที่ต้องการตรวจสอบ

    Returns:
        missing_percentage (float): เปอร์เซ็นต์ข้อมูลที่หายไป (NaN หรือค่าว่าง)
        valid_percentage (float): เปอร์เซ็นต์ข้อมูลที่ไม่หายและตรงกับ Format
        df (pd.DataFrame): DataFrame พร้อมคอลัมน์ Valid ที่บอกว่าข้อมูลผ่านการตรวจสอบหรือไม่
    """

    # ตรวจสอบค่าที่เป็น Missing (NaN หรือค่าว่าง)
    missing_values = df[column_name].isna().sum() + df[column_name].astype(str).apply(lambda x: x.strip() == "").sum()
    missing_percentage = (missing_values / len(df)) * 100

    # ตรวจสอบเฉพาะค่าที่ไม่หายไป (Non-Missing)
    df_valid = df[df[column_name].notna() & (df[column_name].astype(str).str.strip() != "")]
    
    def check_valid_date(value):
        """ ตรวจสอบว่าอยู่ในรูปแบบ YYYY-MM-DD และวันเดือนปีถูกต้อง """
        if not re.fullmatch(r"^\d{4}-\d{2}-\d{2}$", value):
            return False  # ไม่ตรง Format
        
        try:
            year, month, day = map(int, value.split("-"))
            
            # ตรวจสอบปีต้องมากกว่า 1800
            if year < 1800:
                return False

            # ตรวจสอบเดือนต้องอยู่ระหว่าง 1-12
            if month < 1 or month > 12:
                return False

            # ตรวจสอบว่าวันถูกต้องตามเดือนและ Leap Year
            datetime(year, month, day)  # ถ้าวันผิดจะเกิด Error
            
            return True
        except ValueError:
            return False  # ถ้าค่าไม่ถูกต้องตาม Date Format

    # ตรวจสอบค่าที่ถูกต้องตาม Format และ Logic ของวันเดือนปี
    df_valid[f"Valid_{column_name}"] = df_valid[column_name].astype(str).apply(check_valid_date)

    # คำนวณเปอร์เซ็นต์ค่าที่ไม่หายและตรง Format
    total_non_missing = len(df_valid)
    valid_count = df_valid[f"Valid_{column_name}"].sum()
    valid_percentage = (valid_count / total_non_missing) * 100 if total_non_missing > 0 else 0

    # แสดงผลลัพธ์
    print(f"✅ คอลัมน์ '{column_name}' สถานะ:")
    print(f"   - ข้อมูลที่หายไป: {missing_values} จากทั้งหมด {len(df)} ({missing_percentage:.2f}%)")
    print(f"   - ข้อมูลที่ไม่หายและตรง Format: {valid_count} จาก {total_non_missing} ({valid_percentage:.2f}%)")

    return missing_percentage, valid_percentage, df
```

```python
import pandas as pd

def validate_positive_numbers(df, column_name):
    """
    ฟังก์ชันตรวจสอบค่าตัวเลขในคอลัมน์:
    1. คำนวณเปอร์เซ็นต์ของ Missing Values (NaN หรือ ค่าว่าง "")
    2. คำนวณเปอร์เซ็นต์ของค่าที่เป็น Outlier (ค่าที่ <= 0)

    Parameters:
        df (pd.DataFrame): DataFrame ที่ต้องการตรวจสอบ
        column_name (str): ชื่อคอลัมน์ที่ต้องการตรวจสอบ

    Returns:
        missing_percentage (float): เปอร์เซ็นต์ของ Missing Values
        outlier_percentage (float): เปอร์เซ็นต์ของค่าที่เป็น Outlier
        df_missing (pd.DataFrame): DataFrame ที่มีเฉพาะ Missing Values
        df_outliers (pd.DataFrame): DataFrame ที่มีเฉพาะ Outliers
        df (pd.DataFrame): DataFrame พร้อมคอลัมน์ Valid และ Outlier
    """

    # ตรวจสอบค่าที่เป็น Missing (NaN หรือค่าว่าง)
    df["is_missing"] = df[column_name].isna() | df[column_name].astype(str).apply(lambda x: x.strip() == "")
    missing_values = df["is_missing"].sum()
    missing_percentage = (missing_values / len(df)) * 100

    # ตรวจสอบเฉพาะค่าที่ไม่หายไป (Non-Missing) และสร้างสำเนาใหม่
    df_valid = df[~df["is_missing"]].copy()

    # ฟังก์ชันตรวจสอบค่าเป็น Outlier หรือไม่
    def check_valid_number(value):
        try:
            num = float(value)
            return "Outlier" if num <= 0 else "Valid"
        except ValueError:
            return "Outlier"

    # ตรวจสอบ Outliers
    df_valid["Status_Outlier"] = df_valid[column_name].astype(str).apply(check_valid_number)

    # คำนวณเปอร์เซ็นต์ของค่าที่เป็น Outlier
    total_non_missing = len(df_valid)
    outlier_count = (df_valid["Status_Outlier"] == "Outlier").sum()
    outlier_percentage = (outlier_count / total_non_missing) * 100 if total_non_missing > 0 else 0

    # แยก DataFrame สำหรับ Missing และ Outliers
    df_missing = df[df["is_missing"]].copy()
    df_outliers = df_valid[df_valid["Status_Outlier"] == "Outlier"].copy()

    # แสดงผลลัพธ์
    print(f"✅ คอลัมน์ '{column_name}' - Missing Values:")
    print(f"   - ข้อมูลที่หายไป: {missing_values} จากทั้งหมด {len(df)} ({missing_percentage:.2f}%)\n")

    print(f"✅ คอลัมน์ '{column_name}' - Outliers:")
    print(f"   - ข้อมูลที่เป็น Outlier (0 หรือติดลบ): {outlier_count} จาก {total_non_missing} ({outlier_percentage:.2f}%)\n")

    return missing_percentage, outlier_percentage, df_missing, df_outliers, df_valid
```

```python
# ตรวจสอบคอลัมน์ 'Price Per Unit'
missing_price, outlier_price, df_missing, df_outliers, df_validated = validate_positive_numbers(df, "Price Per Unit")
```

```python
def validate_column_values(df, column_name, valid_values):
    """
    ฟังก์ชันตรวจสอบค่าคอลัมน์ว่าตรงกับค่าที่กำหนดกี่เปอร์เซ็นต์ และตรวจสอบ Missing Values
    :param df: DataFrame ที่ต้องการตรวจสอบ
    :param column_name: ชื่อคอลัมน์ที่ต้องการตรวจสอบ
    :param valid_values: รายการค่าที่ถูกต้อง (List)
    :return: DataFrame ที่มีการตรวจสอบ, เปอร์เซ็นต์ค่าที่ถูกต้อง, และเปอร์เซ็นต์ Missing Values
    """
    # ตรวจสอบค่าที่เป็น Missing (NaN หรือ ค่าว่าง "")
    df["is_missing"] = df[column_name].isna() | df[column_name].astype(str).apply(lambda x: x.strip() == "")
    missing_count = df["is_missing"].sum()
    missing_percentage = (missing_count / len(df)) * 100

    # ตรวจสอบว่าค่ามีอยู่ใน List ที่กำหนดหรือไม่ (เฉพาะค่าที่ไม่ใช่ Missing)
    df["Valid"] = df[~df["is_missing"]][column_name].isin(valid_values)

    # คำนวณเปอร์เซ็นต์ค่าที่ถูกต้อง
    total_non_missing = len(df[~df["is_missing"]])
    valid_count = df["Valid"].sum()
    valid_percentage = (valid_count / total_non_missing) * 100 if total_non_missing > 0 else 0

    # แสดงผลลัพธ์
    print(f"✅ คอลัมน์ '{column_name}' สถานะ:")
    print(f"   - ข้อมูลที่หายไป: {missing_count} จากทั้งหมด {len(df)} ({missing_percentage:.2f}%)")
    print(f"   - ข้อมูลที่ถูกต้อง: {valid_count} จาก {total_non_missing} ({valid_percentage:.2f}%)")

    return df[[column_name, "Valid", "is_missing"]], valid_percentage, missing_percentage
```
